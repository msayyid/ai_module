```python
# Step 2: Define path cost
# We did not consider cost when designing the agent last week. In fact, when agent performs
# an action in real world, there is some cost. For example, cost for moving a car is time and
# fuel, and they are correlated with the distance.
# Now, develop some functions to define the cost for your vacuum cleaner. You are suggested
# to define the cost for all the actions that you defined last week.

initial_state = {}
world_state = {}

for i in range(1, 10):
    if i == 1 or i == 2 or i == 3:
        initial_state[i] = 1
        world_state[i] = 1
    else: 
        initial_state[i] = 0
        world_state[i] = 0
# 0 - clean
# 1 - dirty
print("initial state:", initial_state)
print("actual world:", world_state)


goal_state = {}
for i in range(1, 10):
    goal_state[i] = 0
print("goal state:", goal_state)

# current_state = 1
def action(world_state, current_state):
    if world_state[current_state] == 1: # if dirty:
        return "clean" # clean
    elif current_state < 10:
        # move right while you can
        return "move_right"
    else:
        return "stop"

# act = action(initial_state, current_state)
def result(world_state, current_state, act):
    if act == "clean":
        world_state[current_state] = 0 # 0 means cleaned
    elif act == "move_right":
        current_state += 1
    else:
        return -1
    
    return world_state, current_state


# define the cost 
# "clean" costs 3; "move_right" costs 1; "stop" costs 0


def cost_of_act(action):
    if action == "clean":
        return 3

    elif action == "move_right" or action == "move_left":
        return 1
    
    else: return 0

total_cost = 0
current_state = 1
while world_state != goal_state:

    act = action(world_state, current_state)
    cost = cost_of_act(act)
    total_cost += cost
    print("this is the cost variable: ", cost)
    print("this is total cost: ", total_cost)

    if act == "stop":
        break
    
    world_state, current_state = result(world_state, current_state, act)
    print("action:", act, "| state:", current_state, "| world:", world_state)
    print("--------------------------------------------------------------------------------------")
    print("total cost: ", total_cost)

    print("changed")






```
