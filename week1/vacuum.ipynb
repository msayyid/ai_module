```python
# create environment: first 3 dirty, rest clean
environment = ["dirty" if i < 3 else "clean" for i in range(9)]
print(environment)

def vacuum_cleaner(environment):
    for i in range(len(environment)):
        environment[i] = "clean"

vacuum_cleaner(environment)
print(environment)
```

```python
vacuum_cleaner(environment)
print(environment)
```
