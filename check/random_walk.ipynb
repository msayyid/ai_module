```python
import random
import matplotlib.pyplot as plt

# number of steps
steps = 200

x, y = 0, 0
x_path = [x]
y_path = [y]

for _ in range(steps):
    direction = random.choice(["up", "down", "left", "right"])

    if direction == "up":
        y += 1
    elif direction == "down":
        y -= 1
    elif direction == "left":
        x -= 1
    elif direction == "right":
        x += 1

    x_path.append(x)
    y_path.append(y)

# plot
plt.figure()
plt.plot(x_path, y_path, marker="o", markersize=2)
plt.scatter(0, 0, label="Start")
plt.scatter(x, y, label="End")
plt.legend()
plt.title("2D Random Walk")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid()
plt.show()
```
